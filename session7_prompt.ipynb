```python
Date: "10/30/24"
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.arange(0, 1.01, 0.01)

def exp_func():
    return np.exp(x)
```

```python
def main():
    y = exp_func()
    plt.plot(x,y,'b-')
    plt.xlabel('Time [Milliseconds]')
    plt.ylabel('Awesomeness')
    plt.show
    plt.savefig('exp_graph.pdf',bbox_inches='tight',dpi=1000)
    
if __name__=='__main__':
    main()
```
