```python
Date: "10/30/2024"
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
rng = np.random.default_rng()
a = 8.
samples = 1000
s = rng.power(a, samples)
```

```python
width = 0.01
histmin = np.floor(min(s))
histmax = np.ceil(max(s)) + width
bins = np.arange(histmin, histmax, width)
plt.hist(s,bins=bins,alpha=0.5,edgecolor='black')
plt.ylabel('# of times number repeat in power distribution')
plt.xlabel('Randomly generated numbers in power distribution')

plt.savefig('random_power_hist.pdf', bbox_inches='tight', format='pdf', dpi=1000)
```
