```python
Date: "10/30/2024"
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
from numpy import random
```

```python
x = random.rand(1000)

width = 0.01
histmin = np.floor(min(x))
histmax = np.ceil(max(x)) + width

bins = np.arange(histmin,histmax,width)
plt.hist(x,bins=bins,alpha=0.5,edgecolor='black',color='red')
plt.ylabel('# of times numbers (approximately) repeat')
plt.xlabel('Randomly generated numbers')

plt.savefig('random_histogram.pdf', bbox_inches='tight', format='pdf', dpi=1000)
```
