```python
Date: "10/18/2024"
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
def sin_graph(a):
    pi = np.pi
    x = np.arange(0, 2*pi, 0.001)
    y = np.sin(x)
    plt.plot(x,y)
    plt.xlabel("x")
    plt.ylabel(a)
    plt.show
    
if __name__=="__main__":
    sin_graph('sin(x)')
```

First, we must import both numpy and matplotlib

We then define a function (with an argument) in which the x-axis will go from 0 to 2pi in increments of 0.001 using numpy.arange

The argument will allow us to recall the function in the third cell to label the y-axis correctly.

We set y as the sin of x and label the axis accrodingly

Lastly, we initalize the code using the if __name__=="__main__" statement

```python
def cos_graph(b):
    pi = np.pi
    x = np.arange(0, 2*pi, 0.001)
    y = np.cos(x)
    plt.plot(x,y)
    plt.xlabel("x")
    plt.ylabel(b)
    plt.show()
    
if __name__=="__main__":
    cos_graph('cos(x)')
```

Following the same logic as above, we define x as going from 0 to 2 pi with 1000 points in between for more precision.

The function should still have an argument to be able to label the correct y-axis when it is recalled later in cell 3.

We set the y-axis as cos of x and label the x and y aixs accordingly.

Once again, initialize the function.

```python
def cos_and_sin_graph():
    pi = np.pi
    sin_graph('y')
    cos_graph('y')
    plt.show()
    
if __name__=="__main__":
    cos_and_sin_graph()
```
