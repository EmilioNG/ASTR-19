```python
Date: "10/30/2024"
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.arange(0, 1.01, 0.01)

def sin_func():
    return np.sin(x)
    
def cos_func():
    return np.cos(x)
```

```python
f, axarr = plt.subplots(1, 2)

y = sin_func()
z = cos_func()
#first (sin) panel
sin_graph = axarr[0].plot(x, y)
axarr[0].set_xlabel('x')
axarr[0].set_ylabel('sin(x)')
axarr[0].set_title(r'$\sin(x)$')

#second (cos) panel
cos_graph = axarr[1].plot(x,z)
axarr[1].set_xlabel('x')
axarr[1].set_ylabel('cos(x)')
axarr[1].set_title(r'$\cos(x)$')

f.subplots_adjust(wspace=0.3)

plt.savefig('sin_graph.pdf', bbox_inches='tight', format='pdf', dpi=1000)
plt.savefig('cos_graph.pdf', bbox_inches='tight', format='pdf', dpi=1000)
```
